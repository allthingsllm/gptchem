```python
import pandas as pd 
import numpy as np

from gptchem.plotsettings import *
from gptchem.settings import (
    ONE_COL_GOLDEN_RATIO_HEIGHT_INCH,
    ONE_COL_WIDTH_INCH,
    TWO_COL_GOLDEN_RATIO_HEIGHT_INCH,
)
```

```python
gtp3_results = pd.read_pickle('results.pkl')
```

```python
gtp3_results
```

```python
grouped_res_gptj = pd.read_pickle("photoswitch_gptj.pkl")
```

```python
grouped_res_gptj
```

```python
fig, ax = plt.subplots(4,4, figsize=(TWO_COL_GOLDEN_RATIO_HEIGHT_INCH, ONE_COL_WIDTH_INCH*2), sharex='all', sharey='all')

for i, metric in enumerate(['accuracy', 'f1_macro', 'f1_micro', 'kappa']):
    for j, num_epochs in enumerate([4, 8, 14, 20]):
        for k, representation in enumerate(["SMILES", "name", 'selfies']):
            ax[i][j].plot(
            grouped_res_gptj.loc[num_epochs, representation, :].index,
            grouped_res_gptj.loc[num_epochs, representation, :][metric]['mean'],
            label=representation if representation != 'selfies' else 'SELFIES',
            marker='o'
        )
            ax[i][j].fill_between(grouped_res_gptj.loc[num_epochs, representation, :].index,
                                  grouped_res_gptj.loc[num_epochs, representation, :][metric]['mean'] - grouped_res_gptj.loc[num_epochs, representation, :][metric]['sem'],
                                    grouped_res_gptj.loc[num_epochs, representation, :][metric]['mean'] + grouped_res_gptj.loc[num_epochs, representation, :][metric]['sem'],
                                    alpha=0.2
                                    )
            
        ax[i][j].plot(
            gtp3_results.loc[2, "SMILES", :].index,
            gtp3_results.loc[2, "SMILES", :][metric]['mean'],
            label='GPT-3 (SMILES)',
            marker='o'
        )

        ax[i][j].fill_between(gtp3_results.loc[2, "SMILES", :].index,
                                gtp3_results.loc[2, "SMILES", :][metric]['mean'] - gtp3_results.loc[2, "SMILES", :][metric]['sem'],
                                gtp3_results.loc[2, "SMILES", :][metric]['mean'] + gtp3_results.loc[2, "SMILES", :][metric]['sem'],
                                alpha=0.2
                                )
        
        range_frame(ax[i][j], np.array([10,200]), np.array([0, 1])) #+ grouped_res_gptj.loc[num_epochs, representation, :][metric]['sem'].max()
        ax[0][j].set_title(f"{num_epochs} epochs")

        
matplotx.line_labels(ax[0][-1])
# ylabel_top('accuracy')
fig.text(0.33,0.12, 'number of training examples')

ylabel_top('accuracy', ax[0][0])
ylabel_top(r'F$_1$ macro', ax[1][0])
ylabel_top(r'F$_1$ micro', ax[2][0])
ylabel_top(r'$\kappa$', ax[3][0])

# fig.tight_layout()
plt.subplots_adjust(hspace=0.6, top=1, bottom=0.2)

fig.savefig('photoswitch_gptj.pdf', bbox_inches='tight')
```
